## Requête SQL

Montant des ventes le plus faible

```python
query="""
SELECT MIN("MontantTotal") AS "CAHT le plus faible"
FROM sales_fact;
"""
pd.read_sql(sql=query, con=engine)
```

Montant des ventes le plus important

```python
query="""
SELECT MAX("MontantTotal") AS "CAHT le plus important"
FROM sales_fact;
"""
pd.read_sql(sql=query, con=engine)
```

Le produit le plus cher et le produit le moins cher de la table produits

```python
query="""
SELECT 
    MIN("PrixUnitaire") AS "Prix minimum"
    , MAX("PrixUnitaire") AS "Prix maximum"
FROM products_dim;
"""
pd.read_sql(sql=query, con=engine)
```
