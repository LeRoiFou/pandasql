## Requêtes SQL

### Requête n° 1
Donner le nom, le prénom et la moyenne des ventes par clients et conserver les clients pour lesquels aucune vente n'a été réalisée : remplacer les valeurs nulles par 0

````python
query = """
-- Colonnes récupérées
SELECT
    c."Nom"
    , c."Prenom"
    , COALESCE(ROUND(AVG(s."MontantTotal")), 0) AS "Ventes moyennes"
-- BD récupérées
FROM customers_dim c
LEFT JOIN sales_fact s
ON c."ClientID" = s."ClientID"
-- Agrégation
GROUP BY
    c."Nom"
    , c."Prenom"
-- Trie
ORDER BY
    "Ventes moyennes" ASC
    , "Nom" ASC;
"""
pd.read_sql(sql=query, con=engine)
````

### Requête n° 2
Donner le nom, le prénom et la somme des achats réalisés par chaque client, et afficher la valeur zéro dans le cas où certains clients n'ont effectué aucun achat

````python
query = """
-- Colonnes récupérées
SELECT
    c."Nom"
    , c."Prenom"
    , COALESCE(ROUND(SUM(s."MontantTotal")), 0) AS "Total des achats"
-- BD récupérées
FROM customers_dim c
LEFT JOIN sales_fact s
ON c."ClientID" = s."ClientID"
-- Agrégation
GROUP BY
    c."Nom"
    , c."Prenom"
-- Trie
ORDER BY
    "Total des achats" ASC
    , "Nom" ASC;
"""
pd.read_sql(sql=query, con=engine)
````
