Les fonctions d'agrégation récupèrent un ensemble de valeurs pour en restituer une seule
L'instruction DISTINCT() permet de récupérer les valeurs uniques
 

> Syntaxe:
> ```
> SELECT Aggregation(DISTINCT(Colonne)) 
> FROM table;
> ```
> <br>

````python
import duckdb
from skimpy import skim
from skrub import TableReport
````

````python
# Connection à la librairie duckdb
conn = duckdb.connect()
````

## Fichier des ventes

````python
# Requête SQL : récupération de toutes les données du fichier .csv en DF
sales_df = conn.execute(
    """
    SELECT *
    FROM read_csv_auto('data/indexE/Ventes.csv', header=True);
    """
).df()
````

````python
# Stats de la DF ci-avant
skim(sales_df)
````

````python
TableReport(sales_df)
````

````python
# Enregistrement de la DF ci-avant en BD virtuelle (requête SQL : VIEW)
conn.register('sales_db', sales_df)
````

## Requêtes SQL

Montant total du CAHT en prenant les valeurs uniques

````python
conn.execute(
    """
    -- Colonnes récupérées
    SELECT SUM(DISTINCT(MontantTotal)) AS "Valeurs uniques"
    -- BD récupérée
    FROM sales_db;
    """
).df()
````

Montant moyen du CAHT en prenant les valeurs uniques

````python
conn.execute(
    """
    -- Colonnes récupérées
    SELECT AVG(DISTINCT(MontantTotal)) AS "Valeurs uniques"
    -- BD récupérée
    FROM sales_db;
    """
).df()
````

Montant max du CAHT en prenant les valeurs uniques

````python
conn.execute(
    """
    -- Colonnes récupérées
    SELECT MAX(DISTINCT(MontantTotal)) AS "Valeurs uniques"
    -- BD récupérée
    FROM sales_db;
    """
).df()
````

Montant min CAHT en prenant les valeurs uniques

````python
conn.execute(
    """
    -- Colonnes récupérées
    SELECT MIN(DISTINCT(MontantTotal)) AS "Valeurs uniques"
    -- BD récupérée
    FROM sales_db;
    """
).df()
````
