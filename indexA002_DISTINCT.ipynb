> Syntaxe :
> ```
> SELECT DISTINCT nom_colonne
> FROM nom_table;
> ```

````python
import pandas as pd
from skimpy import skim
from sqlalchemy import create_engine
````

## Fichier des ventes

````python
query = """
-- Colonnes récupérées
SELECT *
-- BD récupérée
FROM sales_fact;
"""
sales_df = pd.read_sql(sql=query, con=engine)
sales_df.head(2).transpose()
````

## Requêtes SQL

Donner les différentes dates auxquelles des ventes ont été réalisées

````python
query = """
-- Colonnes récupérées
SELECT DISTINCT("DateVente") AS "Date des ventes"
-- BD récupérée
FROM sales_fact;
"""
pd.read_sql(sql=query, con=engine)
````

Donner les noms et prénoms distincts des employés de l'entreprise

````python
query = """
-- Colonnes récupérées
SELECT
    DISTINCT("Nom")
    , "Prenom"
-- BD récupéréee
FROM employes_dim;
"""
pd.read_sql(sql=query, con=engine)
````
