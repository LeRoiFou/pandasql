## Requêtes SQL

##### Donner pour chaque vente, le nom et le prénom de l'employé qui réalisé la vente

```python
query = """
SELECT
    s."VenteID" AS "N° de vente"
    , s."QuantiteVendue" AS "Quantite vendue"
    , s."MontantTotal" AS "Total vente"
    , e."EmployeID" AS "N° employe"
    , e."Nom"
    , e."Prenom"
FROM sales_fact s
JOIN employes_dim e
ON s."EmployeID" = e."EmployeID";
"""
pd.read_sql(sql=query, con=engine)
```

##### Donner pour chaque produit de la base de données le nom, l'adresse et le n° de téléphone de son fournisseur

```python
query = """
SELECT 
    p.*
    , s."NomFournisseur" AS "Nom"
    , s."Adresse"
    , s."NumeroTelephone" AS "Numero telephone"
FROM products_dim p
JOIN suppliers_dim s
ON p."FournisseurID" = s."FournisseurID"
"""
pd.read_sql(sql=query, con=engine)
```

##### Donner le nom et prénom des employés ayant réalisé la somme des ventes les plus élevées

```python
query = """
SELECT 
    e."EmployeID" AS "N° employe"
    , e."Nom"
    , e."Prenom"
    , ROUND(SUM(s."MontantTotal")) AS "Total ventes"
FROM employes_dim e
JOIN sales_fact s
ON e."EmployeID" = s."EmployeID"
GROUP BY
    "N° employe"
    , e."Nom"
    , e."Prenom"
ORDER BY
    "Total ventes" DESC
    , e."Nom"
    , e."Prenom"
LIMIT 5;
"""
pd.read_sql(sql=query, con=engine)
```

##### Donner pour chaque client, le nom, le prénom, l'adresse ainsi que le nombre d'achats réalisé

```python
query = """
SELECT
    c."ClientID" AS "N° client"
    , c."Nom"
    , c."Prenom"
    , COUNT(s."VenteID") AS "Nombre de ventes"
FROM customers_dim c
JOIN sales_fact s
ON c."ClientID" = s."ClientID"
GROUP BY
    "N° client"
    , c."Nom"
    , c."Prenom"
ORDER BY
    "Nombre de ventes" DESC
    , "N° client"
LIMIT 10;
"""
pd.read_sql(sql=query, con=engine)
```
