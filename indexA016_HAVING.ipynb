## Requêtes SQL

Liste des employés qui réalisent chaque année une moyenne du CAHT > à 1 000 € 

```python
query = """
SELECT
    CAST("EmployeID" AS TEXT) AS "N° employe"
    , CAST(DATE_PART('year', CAST("DateVente" AS DATE)) AS TEXT) AS "Annee"
    , ROUND(AVG("MontantTotal")) AS "CAHT moyen"
FROM sales_fact
GROUP BY
    "N° employe"
    , "Annee"
HAVING ROUND(AVG("MontantTotal")) > 1000
ORDER BY
    "Annee" ASC
    , "CAHT moyen" DESC;
"""

pd.read_sql(sql=query, con=engine)
```
